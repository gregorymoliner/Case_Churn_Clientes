# Analisando Dados com Python

### Case - Cancelamento de Clientes





```python
# Passo 1: Importar a base de dados
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler
```

```python
# Passo 2: Visualizar a base de dados
df = pd.read_csv("cancelamentos.csv")
display(df)
```

```python
# Passo 3: Excluir colunas que nÃ£o ajudam
df = df.drop(columns="CustomerID")
```

```python
display(df)
```

```python
display(df.info())
```

```python
# Passo 4: Excluir valores vazios
df = df.dropna()
```

```python
display(df.info())
```

```python
# Passo 5: AnÃ¡lise inicial. Quantas pessoas cancelaram
cancelamento = df['cancelou'].value_counts()
display(cancelamento)

# Em porcentagem
cancelamento_p = df['cancelou'].value_counts(normalize=True)
display(cancelamento_p)
```

```python
barras = cancelamento.plot(kind='bar', color=['lightcoral', 'lightblue'])

for i in barras.containers:
    barras.bar_label(i, label_type='edge', fontsize=10)
    
plt.title('Cancelamentos', fontsize=16)
plt.xticks(rotation=0, fontsize=10)
plt.ylabel('Quantidade', fontsize=10)
plt.yticks([])
plt.tick_params(axis='x', length=0)
plt.box(False)
plt.grid(False)

plt.show()
```

```python
cores = plt.get_cmap('Pastel1').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.pie(cancelamento_p, labels=cancelamento_p.index, autopct='%1.1f%%')

plt.title('Cancelamentos em %', fontsize=16)

plt.show()
```

```python
# % de Contratos
contagem_contratos = df['duracao_contrato'].value_counts(normalize=True)
display(contagem_contratos)

# Veja que a temos uma divisÃ£o quase igual entre os planos anual e trimestral, mas o plano mensal jÃ¡ fica atrÃ¡s com quase 20%.
```

```python
cores = plt.get_cmap('Pastel1').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.pie(contagem_contratos, labels=contagem_contratos.index, autopct='%.1f')

plt.title('% de Contratos', fontsize=16)

plt.show()
```

```python
barras = cancelamento_p.plot(kind='pie', autopct='%.1f')

plt.title('Cancelamentos em %', fontsize=16)
plt.ylabel('%', fontsize=0)

plt.show()
```

```python
display(df.groupby("duracao_contrato").mean(numeric_only=True))

# Com as informaÃ§Ãµes agrupadas, Ã© possÃ­vel notar que os clientes do plano Mensal, possuem uma mÃ©dia de cancelamento igual a 1, ou seja, praticamente todos os clientes que utilizam esse plano fizeram o cancelamento do serviÃ§o.
```

```python
atraso_cancelamento = df.groupby('dias_atraso')['cancelou'].sum()
```

```python
plt.figure(figsize=(10,6))

plt.hist(df[df['cancelou'] == 0]['dias_atraso'], bins=20, alpha=0.7, label='NÃ£o Cancelou', color='green')
plt.hist(df[df['cancelou'] == 1]['dias_atraso'], bins=20, alpha=0.7, label='Cancelou', color='red')

plt.title('DistribuiÃ§Ã£o dos Dias de Atraso por Cancelamento')
plt.xlabel('Dias de Atraso')
plt.ylabel('FrequÃªncia')
plt.legend()
plt.grid(True)

plt.show()

# Podemos verificar que apÃ³s 20 dias de atraso temos pouca sobreposiÃ§Ã£o, significando que os clientes que cancelaram e os que nÃ£o cancelaram tiveram comportamentos diferentes.
# Isso pode indicar que o atraso estÃ¡ fortemente relacionado ao cancelamento.
# Se as barras vermelhas (cancelamento) estiverem concentradas em dias de atraso mais altos, isso indicaria que quanto maior o atraso, maior chance de cancelamento.

# ConclusÃ£o
# Os cancelamentos estÃ£o mais concentrados nos clientes com mais de 20 dias de atraso, pode ser um sinal de que atrasos contribuem significativamente para o cancelamento.


```

```python
plt.figure(figsize=(10,6))

plt.hist(df[df['cancelou'] == 0]['ligacoes_callcenter'], bins=20, alpha=0.7, label='NÃ£o Cancelou', color='green')
plt.hist(df[df['cancelou'] == 1]['ligacoes_callcenter'], bins=20, alpha=0.7, label='Cancelou', color='red')

plt.title('DistribuiÃ§Ã£o dos Dias de Atraso por Call Center')
plt.xlabel('Dias de Atraso')
plt.ylabel('FrequÃªncia')
plt.legend()
plt.grid(True)

plt.show()

# Podemos verificar que clientes com mais de 5 ligaÃ§Ãµes ao call center cancelam suas assinaturas.
```

```python
# Com os grÃ¡ficos conseguimos analisar que acima de 20 dias, 100% efetua o cancelamento e ligaÃ§aÃµes call center acima de 5 todos cancelam.
tabela = df[df["ligacoes_callcenter"]<5]
tabela = df[df["dias_atraso"]<=20]
tabela = df[df["duracao_contrato"]!="Monthly"]
display(tabela)

display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))

# Mantendo na tabela as ligaÃ§Ãµes ao call center que sÃ£o menores do que 5 e os dias menores ou iguais a 20.
# Reduzimos os cancelamentos de quase 50% para 18.4%.
```
